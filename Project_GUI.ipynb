```python
from tkinter import *
from tkinter import messagebox
import sqlite3


#Connecting and Creating Database
conn = sqlite3.connect('Sample.db')
cursor=conn.cursor()
cursor.execute('CREATE TABLE IF NOT EXISTS Employe (Username TEXT NOT NULL,Password TEXT NOT NULL,Name TEXT NOT NULL,Email TEXT NOT NULL,Location TEXT NOT NULL,Department TEXT NOT NULL)')
conn.commit()
 
def savedb(): # Saving data to database
    
    cursor.execute('INSERT INTO Employe (Username,Password,Name,Email,Location,Department) VALUES(?,?,?,?,?,?)',(username_entry.get(),password_entry.get(),name_entry.get(),email_entry.get(),loc_entry.get(),dept_entry.get()))
    conn.commit()
    
    if(username.get() == ""):
        messagebox.showinfo('Information','Please enter Username')
    elif(password.get() == ""):
        messagebox.showinfo('Information','Please enter a Password')
    elif(con_password.get() == ""):
        messagebox.showinfo('Information','Please enter a Password')
    elif(password.get() != con_password.get()):
        messagebox.showinfo('Information','Passwords doesnot matches')
    elif(Name.get() == ""):
        messagebox.showinfo('Information','Please enter a valid Name')
    elif(Email.get() == ""):
        messagebox.showinfo('Information','Please enter valid Email')
    elif(Location.get() == ""):
        messagebox.showinfo('Information','Please enter Location')
    elif(department.get() == ""):
        messagebox.showinfo('Information','Please enter your Department')
    else:
        messagebox.showinfo('Information','Registration successful')
        username_entry.delete(0,END)
        password_entry.delete(0,END)
        con_password_entry.delete(0,END)
        name_entry.delete(0,END)
        email_entry.delete(0,END)
        loc_entry.delete(0,END)
        dept_entry.delete(0,END)
    
def verify_login(): 
    
    find_emp = ("SELECT * FROM Employe WHERE Username = ? and Password = ?")
    cursor.execute(find_emp,(Username.get(),Password.get()))
    result = cursor.fetchall()
    
    if(Username.get() == ""):
        messagebox.showinfo('Information','Please enter Username')
    elif(Password.get() == ""):
        messagebox.showinfo('Information','Please enter a Password')
    elif result:
        print(result)
        messagebox.showinfo('Information','Login Success')
        userlogin_entry.delete(0,END)
        pwdlogin_entry.delete(0,END)
    else:
        messagebox.showinfo('Information','Invalid Username/Password')

#Creating window
root = Tk()
root.title("Employee Login/Registation")
root.state("zoomed")

#Creating frame for Login

f1 = Frame(root,highlightthickness=4,bg="teal",borderwidth=4,relief="groove",height=220,width=350)
f1.place(x=300,y=150)


l1=Label(f1,text="Enter your details below to Login",bg="white",font=("calibri",15))
l1.place(x=30,y=20)

Username = StringVar()
Password = StringVar()

usernameLabel = Label(f1, text="Username",bg="white")
usernameLabel.place(x=60,y=80)
userlogin_entry = Entry(f1, textvariable=Username,bg="white")
userlogin_entry.place(x=140,y=80,height=20)

#password label and password entry box
passwordLabel = Label(f1,text="Password",bg="white")
passwordLabel.place(x=60,y=120)

pwdlogin_entry = Entry(f1, textvariable=Password, show='*',bg="white")
pwdlogin_entry.place(x=140,y=120)

loginButton = Button(f1, text="Login", command=verify_login)
loginButton.place(x=160,y=160,width=70,height=30)

l3 = Label(root,text="-- Or --",font=("calibri",15))
l3.place(x=700,y=250)

#Employee Registration deatils

username = StringVar()
password = StringVar()
con_password = StringVar()
Name = StringVar()
Email = StringVar()
Location = StringVar()
department = StringVar()

#Creating frame for Registration
f2 = Frame(root,bg="teal",highlightthickness=6,borderwidth=4,relief="groove",height=500,width=400)
f2.place(x=800,y=50)

l2 = Label(f2,text="Not yet Regsitered?\nPlease enter details below to Register",bg="white",font=("calibri",15))
l2.place(x=40,y=20)

username_lable = Label(f2,text="Username",bg="white")
username_lable.place(x=90,y=100)
username_entry = Entry(f2,textvariable=username,bg="white")
username_entry.place(x=170,y=100)

password_lable = Label(f2, text="Password",bg="white")
password_lable.place(x=90,y=150)
password_entry = Entry(f2, textvariable=password, show='*',bg="white")
password_entry.place(x=170,y=150)

con_password_lable = Label(f2, text="Confirm Password",bg="white")
con_password_lable.place(x=60,y=200)
con_password_entry = Entry(f2, textvariable=con_password, show='*',bg="white")
con_password_entry.place(x=170,y=200)

name_lable = Label(f2, text="Name",bg="white")
name_lable.place(x=90,y=250)
name_entry = Entry(f2, textvariable=Name,bg="white")
name_entry.place(x=170,y=250)

email_lable = Label(f2, text="Email",bg="white")
email_lable.place(x=90,y=300)
email_entry = Entry(f2,textvariable=Email,bg="white")
email_entry.place(x=170,y=300)

loc_lable = Label(f2, text="Location",bg="white")
loc_lable.place(x=90,y=350)
loc_entry = Entry(f2, textvariable=Location,bg="white")
loc_entry.place(x=170,y=350)

dept_lable = Label(f2, text="Department",bg="white")
dept_lable.place(x=90,y=400)
dept_entry = Entry(f2, textvariable=department,bg="white")
dept_entry.place(x=170,y=400)

registerButton = Button(f2, text="Submit",command=savedb)
registerButton.place(x=180,y=440,width=70,height=30)

root.mainloop()
```
